```python
#import necessary library
import numpy as np
import matplotlib.pyplot as plt

#generate random data
#variable "a" : 100 random values between 0 and 1
a = np.random.rand(100)

#Variable "b" : linear relationship with "a" plus some random noise
b = 5 + 3*a + np.random.normal(size=100)


#create a scatter plot to visualize the data 
plt.scatter(a, b, color = 'red')


#Add labels and title to the plot
plt.title("Linear relationship between 'a' and 'b' with noise")
plt.xlabel("Predictor 'a'")
plt.ylabel("Response 'b'")

#show the plot
plt.show()
```
